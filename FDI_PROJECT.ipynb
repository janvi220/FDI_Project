CLEANING DATA:

```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\Admin\FDI_Project\FDI\fdi_rawdata.csv"
df = pd.read_csv(file_path)

# Step 1: Remove completely empty columns
df = df.dropna(axis=1, how='all')

# Step 2: Remove rows where all yearly values are missing
year_columns = [col for col in df.columns if col.isdigit()]  # Get year columns
df = df.dropna(subset=year_columns, how='all')

# Step 3: Remove duplicate rows if any
df = df.drop_duplicates()

# Step 4: Trim spaces from column names
df.columns = df.columns.str.strip()

# Save the cleaned dataset
cleaned_file_path = "C:\\Users\\Admin\\FDI_Project\\FDI\\fdi_cleaningdata.csv"
df.to_csv(cleaned_file_path, index=False)

print(f"Cleaned dataset saved at: {cleaned_file_path}")
```

```python
import pandas as pd

# Load the data
df = pd.read_csv("C:\\Users\\Admin\\FDI_Project\\FDI\\fdi_cleaningdata.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Melt the DataFrame to transform year columns into rows
melted_df = df.melt(id_vars=['Country Name'],
                    value_vars=[str(year) for year in range(2022, 2024)],  # Adjust year range as needed
                    var_name='Year',
                    value_name='Sales')

# Filter conditions
country_filter = ['India']  # Add desired countries
year_filter = ['2022', '2023']      # Years must be strings

# Extract data by Country and Year
filtered_df = melted_df[(melted_df['Country Name'].isin(country_filter)) &
                        (melted_df['Year'].isin(year_filter))]

# Display the filtered data
print(filtered_df)

# Optional: Save filtered data to a CSV
filtered_df.to_csv(r"C:\Users\Admin\FDI_Project\FDI\filtered_sales_data.csv", index=False)
```

2.TRANSFORM

```python
import pandas as pd

# Load the filtered data
filtered_df = pd.read_csv(r"C:\Users\Admin\FDI_Project\FDI\filtered_sales_data.csv")

# === DATA CLEANING ===
# Handle missing values by filling with 0
filtered_df = filtered_df.assign(Sales=filtered_df['Sales'].fillna(0))


# Convert 'Sales' column to float for calculations
filtered_df['Sales'] = pd.to_numeric(filtered_df['Sales'], errors='coerce').fillna(0)

# === DATA AGGREGATION ===
# Group by Country and Year, summing sales
transformed_df = filtered_df.groupby(['Country Name', 'Year']).agg({'Sales': 'sum'}).reset_index()

# === DATA ENRICHMENT ===
# Add a column for percentage growth (Year-over-Year growth)
transformed_df['Sales Growth (%)'] = transformed_df.groupby('Country Name')['Sales'].pct_change().fillna(0) * 100

# === DATA FORMATTING ===
# Rename columns for clarity
transformed_df.rename(columns={'Country Name': 'Country'}, inplace=True)

# Save the transformr"C:\Users\Admin\FDI_Project\FDI\filtered_sales_data.csv"FDI\\transformed_sales_data.csv", index=False)

print("Data transformation complete. File saved as 'transformed_sales_data.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima

# Load dataset
file_path = "C:\\Users\\Admin\\FDI_Project\\ETL_FDI\\fdi_cleaningdata.csv"
df = pd.read_csv(file_path)

# Filter data for United States
df_usa = df[df["Country Name"] == "United States"]

# Convert wide format to long format
df_long = df_usa.melt(id_vars=["Country Name"], var_name="Year", value_name="FDI Inflows")

# Convert Year and FDI Inflows to numeric
df_long["Year"] = pd.to_numeric(df_long["Year"], errors='coerce')
df_long["FDI Inflows"] = pd.to_numeric(df_long["FDI Inflows"], errors='coerce')

# Filter for years 2010-2023
df_long = df_long[(df_long["Year"] >= 2010) & (df_long["Year"] <= 2023)]

# Set Year as index
df_long.set_index("Year", inplace=True)

# Fill missing values (if any)
df_long["FDI Inflows"].fillna(method="ffill", inplace=True)

# Plot historical data
plt.figure(figsize=(10,5))
plt.plot(df_long.index, df_long["FDI Inflows"], marker="o", label="Historical FDI")
plt.xlabel("Year")
plt.ylabel("FDI Inflows (USD)")
plt.title("FDI Inflows for United States (2010-2023)")
plt.legend()
plt.show()
```

```python
# Auto ARIMA to find the best order (p, d, q)
best_model = auto_arima(df_long["FDI Inflows"], seasonal=False, stepwise=True)
print(best_model.summary())

# Fit ARIMA with best parameters
p, d, q = best_model.order
model = ARIMA(df_long["FDI Inflows"], order=(p, d, q))
model_fit = model.fit()

# Forecast for next 5 years (2024-2028)
future_years = [2024, 2025, 2026, 2027, 2028]
forecast = model_fit.forecast(steps=5)

# Create forecast dataframe
forecast_df = pd.DataFrame({"Year": future_years, "Predicted FDI": forecast})

# Plot forecast results
plt.figure(figsize=(10,5))
plt.plot(df_long.index, df_long["FDI Inflows"], marker="o", label="Historical FDI")
plt.plot(future_years, forecast, marker="o", linestyle="dashed", color="red", label="Predicted FDI")
plt.xlabel("Year")
plt.ylabel("FDI Inflows (USD)")
plt.title("FDI Forecast for United States (2024-2028)")
plt.legend()
plt.show()

# Print forecasted FDI values
print("🔮 Predicted FDI Inflows (2024-2028):")
print(forecast_df)
```
